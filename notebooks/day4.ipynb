```python
text = open('../data/day4.txt', 'r').read()

def find_horizontal(text, count):
    lines = [line.strip() for line in text.split('\n') if line.strip()]
    height = len(lines)
    width = len(lines[0])
    
    # Right direction (0,1)
    for i in range(height):
        for j in range(width):
            valid = True
            for offset in range(4):
                new_j = j + offset
                if new_j >= width:
                    valid = False
                    break
                if lines[i][new_j] != "XMAS"[offset]:
                    valid = False
                    break
            if valid:
                count += 1
    
    # Left direction (0,-1)
    for i in range(height):
        for j in range(width):
            valid = True
            for offset in range(4):
                new_j = j - offset
                if new_j < 0:
                    valid = False
                    break
                if lines[i][new_j] != "XMAS"[offset]:
                    valid = False
                    break
            if valid:
                count += 1
    return count

def find_vertical(text, count):
    lines = [line.strip() for line in text.split('\n') if line.strip()]
    height = len(lines)
    width = len(lines[0])
    
    # Down direction (1,0)
    for i in range(height):
        for j in range(width):
            valid = True
            for offset in range(4):
                new_i = i + offset
                if new_i >= height:
                    valid = False
                    break
                if lines[new_i][j] != "XMAS"[offset]:
                    valid = False
                    break
            if valid:
                count += 1
    
    # Up direction (-1,0)
    for i in range(height):
        for j in range(width):
            valid = True
            for offset in range(4):
                new_i = i - offset
                if new_i < 0:
                    valid = False
                    break
                if lines[new_i][j] != "XMAS"[offset]:
                    valid = False
                    break
            if valid:
                count += 1
    return count

def find_diagonal_right(text, count):
    lines = [line.strip() for line in text.split('\n') if line.strip()]
    height = len(lines)
    width = len(lines[0])
    
    # Down-right direction (1,1)
    for i in range(height):
        for j in range(width):
            valid = True
            for offset in range(4):
                new_i = i + offset
                new_j = j + offset
                if new_i >= height or new_j >= width:
                    valid = False
                    break
                if lines[new_i][new_j] != "XMAS"[offset]:
                    valid = False
                    break
            if valid:
                count += 1
                
    # Up-left direction (-1,-1)
    for i in range(height):
        for j in range(width):
            valid = True
            for offset in range(4):
                new_i = i - offset
                new_j = j - offset
                if new_i < 0 or new_j < 0:
                    valid = False
                    break
                if lines[new_i][new_j] != "XMAS"[offset]:
                    valid = False
                    break
            if valid:
                count += 1
    return count

def find_diagonal_left(text, count):
    lines = [line.strip() for line in text.split('\n') if line.strip()]
    height = len(lines)
    width = len(lines[0])
    
    # Up-right direction (-1,1)
    for i in range(height):
        for j in range(width):
            valid = True
            for offset in range(4):
                new_i = i - offset
                new_j = j + offset
                if new_i < 0 or new_j >= width:
                    valid = False
                    break
                if lines[new_i][new_j] != "XMAS"[offset]:
                    valid = False
                    break
            if valid:
                count += 1
                
    # Down-left direction (1,-1)
    for i in range(height):
        for j in range(width):
            valid = True
            for offset in range(4):
                new_i = i + offset
                new_j = j - offset
                if new_i >= height or new_j < 0:
                    valid = False
                    break
                if lines[new_i][new_j] != "XMAS"[offset]:
                    valid = False
                    break
            if valid:
                count += 1
    return count

count = 0
count = find_horizontal(text, count)
count = find_vertical(text, count)
count = find_diagonal_right(text, count)
count = find_diagonal_left(text, count)
print(count)
```

```python
def find_patterns(input_text: str) -> int:
    lines = [line.strip() for line in input_text.split('\n') if line.strip()]
    height = len(lines)
    width = len(lines[0])
    total = 0
    
    # Just checking diagonal directions
    dirs = [(1, 1), (1, -1)]
    
    for i in range(height):
        for j in range(width):
            # Only check if we find an 'A'
            if lines[i][j] != 'A':
                continue
                
            valid = True
            # Check both diagonal directions
            for direction in dirs:
                # Get positions one step in each direction from A
                x1 = i + direction[0]
                y1 = j + direction[1]
                x2 = i - direction[0]
                y2 = j - direction[1]
                
                # Check boundaries
                if (x1 < 0 or x1 >= height or y1 < 0 or y1 >= width or
                    x2 < 0 or x2 >= height or y2 < 0 or y2 >= width):
                    valid = False
                    continue
                    
                # Check if we have M and S in either order
                if ((lines[x1][y1] == 'M' and lines[x2][y2] == 'S') or
                    (lines[x1][y1] == 'S' and lines[x2][y2] == 'M')):
                    valid = True
                    break
                valid = False
                
            if valid:
                total += 1
                
    return total

print(find_patterns(text))
```
