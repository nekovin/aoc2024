```python
text = open('../data/day6.txt').read().split("\n")[:-1]
```

```python

guard_map = [list(row) for row in text]

print(guard_map)

def print_map():
    for row in guard_map:
        print(row)

def get_guard_position():
    for y, row in enumerate(guard_map):
        for x, char in enumerate(row):
            if char in ['^', '>', 'v', '<']:
                return x, y, char

x, y, direction = get_guard_position()
visited = [(x, y)] 
count = 1 

print(get_guard_position())
print_map()

def move_forward(x, y, direction, grid):
    grid[y][x] = 'X'
    if direction == '^':
        grid[y-1][x] = '^'
        return x, y-1, direction
    elif direction == '>':
        grid[y][x+1] = '>'
        return x+1, y, direction
    elif direction == 'v':
        grid[y+1][x] = 'v'
        return x, y+1, direction
    elif direction == '<':
        grid[y][x-1] = '<'
        return x-1, y, direction
    
def detect_obstacle(x, y, direction, grid):
    if direction == '^' and grid[y-1][x] == '#':
        return True
    elif direction == '>' and grid[y][x+1] == '#':
        return True
    elif direction == 'v' and grid[y+1][x] == '#':
        return True
    elif direction == '<' and grid[y][x-1] == '#':
        return True
    return False 

def turn_90_degrees(direction):
    if direction == '^':
        return '>'
    elif direction == '>':
        return 'v'
    elif direction == 'v':
        return '<'
    elif direction == '<':
        return '^'

def check_inbounds(x, y, direction, grid):
    if direction == '^' and y-1 >= 0:
        return True
    elif direction == '>' and x+1 < len(grid[0]):
        return True
    elif direction == 'v' and y+1 < len(grid):
        return True
    elif direction == '<' and x-1 >= 0:
        return True
    else:
        return False

inb = True
while inb == True:
    if not check_inbounds(x, y, direction, guard_map):
        print("Guard has left the map")
        #print_map()
        inb = False
        break

    if not detect_obstacle(x, y, direction, guard_map):
        x, y, direction = move_forward(x, y, direction, guard_map)

        if (x, y) in visited:
            #print("Visited")
            pass
        else:
            visited.append((x, y))
            count += 1
    else:
        direction = turn_90_degrees(direction)

print_map()
print(x, y)
print(count)
```

```python
def get_guard_position(guard_map):
    for y, row in enumerate(guard_map):
        for x, cell in enumerate(row):
            if cell == '^':
                return x, y, '^'
    return None, None, None

def move_forward(x, y, direction):
    dx, dy = {'<': (-1, 0), '>': (1, 0), '^': (0, -1), 'v': (0, 1)}[direction]
    return x + dx, y + dy

def turn_90_degrees(direction):
    turns = {'<': '^', '^': '>', '>': 'v', 'v': '<'}
    return turns[direction]

def check_inbounds(x, y, guard_map):
    return 0 <= y < len(guard_map) and 0 <= x < len(guard_map[0])

def has_loop(start_x, start_y, start_direction, guard_map):
    visited_states = set()
    x, y, direction = start_x, start_y, start_direction
    
    while True:
        state = (x, y, direction)
        if state in visited_states:
            return True
        visited_states.add(state)
        
        next_x, next_y = move_forward(x, y, direction)
    
        if not check_inbounds(next_x, next_y, guard_map):
            return False
            
        if guard_map[next_y][next_x] == '#':
            direction = turn_90_degrees(direction)
        else:
            x, y = next_x, next_y
            
def find_loops(guard_map):
    start_x, start_y, start_dir = get_guard_position(guard_map)
    loop_count = 0

    for y in range(len(guard_map)):
        for x in range(len(guard_map[0])):
            if guard_map[y][x] == '.' and (x, y) != (start_x, start_y):
                test_map = [row[:] for row in guard_map]
                test_map[y][x] = '#'
                
                if has_loop(start_x, start_y, start_dir, test_map):
                    loop_count += 1
                    
    return loop_count

guard_map = [list(row) for row in text]
result = find_loops(guard_map)
print(f"{result}")
```
