```python


guard_map = text.copy()
guard_map = [list(row) for row in text]

print(guard_map)

def print_map():
    for row in guard_map:
        print(row)

def get_guard_position():
    for y, row in enumerate(guard_map):
        for x, char in enumerate(row):
            if char in ['^', '>', 'v', '<']:
                return x, y, char

x, y, direction = get_guard_position()
visited = [(x, y)] 
count = 1 

print(get_guard_position())
print_map()

def move_forward(x, y, direction, grid):
    grid[y][x] = 'X'
    if direction == '^':
        grid[y-1][x] = '^'
        return x, y-1, direction
    elif direction == '>':
        grid[y][x+1] = '>'
        return x+1, y, direction
    elif direction == 'v':
        grid[y+1][x] = 'v'
        return x, y+1, direction
    elif direction == '<':
        grid[y][x-1] = '<'
        return x-1, y, direction
    
def detect_obstacle(x, y, direction, grid):
    if direction == '^' and grid[y-1][x] == '#':
        return True
    elif direction == '>' and grid[y][x+1] == '#':
        return True
    elif direction == 'v' and grid[y+1][x] == '#':
        return True
    elif direction == '<' and grid[y][x-1] == '#':
        return True
    return False 

def turn_90_degrees(direction):
    if direction == '^':
        return '>'
    elif direction == '>':
        return 'v'
    elif direction == 'v':
        return '<'
    elif direction == '<':
        return '^'

def check_inbounds(x, y, direction, grid):
    if direction == '^' and y-1 >= 0:
        return True
    elif direction == '>' and x+1 < len(grid[0]):
        return True
    elif direction == 'v' and y+1 < len(grid):
        return True
    elif direction == '<' and x-1 >= 0:
        return True
    else:
        return False

inb = True
while inb == True:
    if not check_inbounds(x, y, direction, guard_map):
        print("Guard has left the map")
        #print_map()
        inb = False
        break

    if not detect_obstacle(x, y, direction, guard_map):
        x, y, direction = move_forward(x, y, direction, guard_map)

        if (x, y) in visited:
            #print("Visited")
            pass
        else:
            visited.append((x, y))
            count += 1
    else:
        direction = turn_90_degrees(direction)

print_map()
print(x, y)
print(count)
```

```python
class Guard:
    def __init__(self, direction, position):
        self.direction = direction  # (dx, dy)
        self.position = position    # (x, y)
    
    def clone(self):
        return Guard(self.direction, self.position)

def part1(input_text):
    lines = [list(line) for line in input_text.split('\n')]
    height = len(lines)
    width = len(lines[0])
    
    # Find guard starting position
    guard = Guard((0, 1), (0, 0))
    for x in range(height):
        for y in range(width):
            if lines[x][y] == '^':
                guard = Guard((-1, 0), (x, y))

    total = 1
    visited = {guard.position}
    
    while True:
        x = guard.position[0] + guard.direction[0]
        y = guard.position[1] + guard.direction[1]
        
        # Check bounds
        if x < 0 or x >= height or y < 0 or y >= width:
            break
            
        # Handle obstacle
        if lines[x][y] == '#':
            # Turn right
            dx, dy = guard.direction
            if (dx, dy) == (0, 1):   # right
                guard.direction = (1, 0)
            elif (dx, dy) == (0, -1): # left
                guard.direction = (-1, 0)
            elif (dx, dy) == (1, 0):  # down
                guard.direction = (0, -1)
            elif (dx, dy) == (-1, 0): # up
                guard.direction = (0, 1)
                
            x = guard.position[0] + guard.direction[0]
            y = guard.position[1] + guard.direction[1]
            
            if x < 0 or x >= height or y < 0 or y >= width:
                break
        
        if (x, y) not in visited:
            total += 1
            visited.add((x, y))
            
        guard.position = (x, y)
    
    return total

def has_loop(guard, obstructed):
    visited = set()
    height = len(obstructed)
    width = len(obstructed[0])
    
    while True:
        state = (guard.position, guard.direction)
        if state in visited:
            return True
            
        visited.add(state)
        
        x = guard.position[0] + guard.direction[0]
        y = guard.position[1] + guard.direction[1]
        
        if x < 0 or x >= height or y < 0 or y >= width:
            break
            
        while obstructed[x][y] == '#':
            dx, dy = guard.direction
            if (dx, dy) == (0, 1):   # right
                guard.direction = (1, 0)
            elif (dx, dy) == (0, -1): # left
                guard.direction = (-1, 0)
            elif (dx, dy) == (1, 0):  # down
                guard.direction = (0, -1)
            elif (dx, dy) == (-1, 0): # up
                guard.direction = (0, 1)
                
            x = guard.position[0] + guard.direction[0]
            y = guard.position[1] + guard.direction[1]
            
            if x < 0 or x >= height or y < 0 or y >= width:
                return False
                
        guard.position = (x, y)
    
    return False

def part2(input_text):
    lines = [list(line) for line in input_text.split('\n')]
    
    # Find guard starting position
    guard = Guard((0, 1), (0, 0))
    for x in range(len(lines)):
        for y in range(len(lines[0])):
            if lines[x][y] == '^':
                guard = Guard((-1, 0), (x, y))
    
    total = 0
    
    # Try each possible obstacle position
    for x in range(len(lines)):
        for y in range(len(lines[0])):
            if lines[x][y] in ['#', '^']:
                continue
                
            obstructed = [row[:] for row in lines]
            obstructed[x][y] = '#'
            
            if has_loop(guard.clone(), obstructed):
                total += 1
                
    return total

# Test with example
example = """....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#..."""

print(part1(example))  # Should print 41
print(part2(example))  # Should print 6

# Run with your input
with open('../data/day6.txt') as f:
    input_text = f.read().strip()
print(part2(input_text))
```
