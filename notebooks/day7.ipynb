```python
text = open('../data/day7.txt').read().strip().split("\n")

# Parse each line to extract target and nums
parsed_data = []
for line in text:
    target, nums = line.split(':')  # Split the line into target and numbers
    target = int(target.strip())  # Convert the target to an integer
    nums = nums.strip().split(' ')  # Split the numbers into a list
    parsed_data.append((target, nums))  # Append as a tuple (target, nums)

# Display the parsed data
parsed_data
```

```python
targets, nums_list = zip(*parsed_data)
```

```python
operators = ['+', '*']
```

```python
from itertools import product

def evaluate_equ(result):
    first, operator = '', ''
    values = []  
    
    for c in result:
        if c.isdigit():
            first += c  
        elif c in operators:
            values.append(int(first))  
            values.append(c)  
            first = '' 
    values.append(int(first)) 

    result = values[0]
    for i in range(1, len(values), 2):
        operator = values[i]
        next_num = values[i + 1]
        if operator == '+':
            result += next_num
        elif operator == '*':
            result *= next_num
    
    return result

results = []

for nums, target in zip(nums_list, targets):

    #print(nums, target)

    new_equation = '+'.join(nums) 

    operator_positions = [i for i, char in enumerate(new_equation) if char == '+']
    num_operators = len(operator_positions)

    
    for combination in product(operators, repeat=num_operators):
        eq_list = list(new_equation)
        for pos, op in zip(operator_positions, combination):
            eq_list[pos] = op
        generated_equation = ''.join(eq_list) 
        if evaluate_equ(generated_equation) == target:
            results.append(target)
            break

print(results)
print(sum(results))
```
